```python
# Retirando colunas que não serão utilizadas, utilizando função nunique para analisar as colunas
dataset1.nunique()
# Market_place -> possui apenas 1 resultado, ou seja, apenas um valor. Logo todos os produtos são
# de apenas uma localização no caso US(Estados Unidos), tornando-a desnecessária.
# Review_data -> Possui apenas 1 resultado também, logo todos os reviews são da mesma data
# Product_category -> Possui apenas 1 resultado, todos os produtos são da mesma categoria
# product_parent -> Não temos demais informações sobre a sua importância

dataset1_drop = dataset1.drop(columns=['market_place', 'product_id', 'vine', 'review_date', 'product_category', 'product_parent'])
```

```python
# Entendendo coluna por coluna
dataset1_drop.head()
```

```python
#Tirando aspas dos produtos, pois se utilizarmos baseados em seus nomes teremos problemas com as aspas iniciais
def tirando_aspas(value):
    value=str(value)
    value=value.split('"')
    value=value[1]
    return value
dataset1_drop['product_title']=dataset1_drop['product_title'].apply(tirando_aspas)
dataset1_drop.head()
```

```python
dataset1_drop.star_rating.value_counts()
# Nota-se que a coluna de avaliação de 0 a 5, contém apenas os valores 0 e 1.
```

```python
# Tratando caracteres especiais da coluna verified_purchase
dataset1_drop.verified_purchase.value_counts()
#Há apenas valores de 1 \t(Y) e 0 \t(N)  preciso tratar

def lidando_verified_purchase(value):
    value=value[4]
    return value

dataset1_drop['verified_purchase']=dataset1_drop['verified_purchase'].apply(lidando_verified_purchase)
dataset1_drop.head()
```

```python
#Trocando Y e N para Sim e Nao

def y_n_para_sim_nao(value):
    if value=="Y":
        value = "Sim"
    else:
        value = "Nao"
    return value
dataset1_drop['verified_purchase']=dataset1_drop['verified_purchase'].apply(y_n_para_sim_nao)
dataset1_drop.head()
```
